```python
import cartopy
import matplotlib.pyplot as mp
```

```python
crs = cartopy.crs.InterruptedGoodeHomolosine()

fig = mp.figure()
ax = fig.add_axes(111, projection=crs)
ax.add_feature(cartopy.feature.COASTLINE, color='#000', linewidth=1)
ax.spines['geo'].set_edgecolor('#000')
ax.spines['geo'].set_linewidth(1)
fig.tight_layout()
fig.savefig('../docs/homolosine.svg', transparent=True, bbox_inches='tight')
```
